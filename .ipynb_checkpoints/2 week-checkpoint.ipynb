```python
import numpy as np
import cv2
```

```python
## 사진의 이진화

src = cv2.imread('image/iron.jpg', 1)
src = cv2.resize(src, dsize = (0,0),fx = 0.6, fy = 0.6, interpolation = cv2.INTER_AREA)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)

ret1, dst1 = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
ret2, dst2 = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY_INV)
ret3, dst3 = cv2.threshold(gray, 100, 255, cv2.THRESH_TRUNC)
ret4, dst4 = cv2.threshold(gray, 100, 255, cv2.THRESH_TOZERO)
ret5, dst5 = cv2.threshold(gray, 100, 255, cv2.THRESH_TOZERO_INV)
ret6, dst6 = cv2.threshold(gray, 100, 255, cv2.THRESH_MASK)
otsu, dst7 = cv2.threshold(gray, 100, 255, cv2.THRESH_OTSU)
ret8, dst8 = cv2.threshold(gray, 100, 255, cv2.THRESH_TRIANGLE)

print(otsu)
cv2.imshow('gray',gray)
cv2.imshow('binary',dst1)
cv2.imshow('binary_inv',dst2)
cv2.imshow('trunc',dst3)
cv2.imshow('tozero',dst4)
cv2.imshow('tozero_inv',dst5)
cv2.imshow('mask',dst7)
cv2.imshow('ostu',dst7)
cv2.imshow('triangle',dst8)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## 적응형 이진화

src = cv2.imread('image/iron.jpg', 1)
src = cv2.resize(src, dsize = (0,0),fx = 0.65,fy = 0.65, interpolation = cv2.INTER_AREA)
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
binary1 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,3)
binary2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
binary3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C|cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
otsu, dst = cv2.threshold(gray, 100, 255, cv2.THRESH_OTSU)

cv2.imshow('otsu', dst)
cv2.imshow('gray', gray)
cv2.imshow('mean', binary1)
cv2.imshow('gaussian', binary2)
cv2.imshow('mean or gaussian', binary3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 윤곽선 검출

import numpy as np
import cv2

src = cv2.imread('image/contours4.png', cv2.IMREAD_COLOR)
src2 = src.copy()

gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
binary = cv2.bitwise_not(binary)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_TC89_L1)

for i in range(len(contours)):
    cv2.drawContours(src, [contours[i]], 0, (0,0,255),2)
    cv2.putText(src, str(i), tuple(contours[i][0][0]),cv2.FONT_HERSHEY_COMPLEX, 0.8,(0,255,0),1)
    print(i, hierarchy[0][i])
    cv2.imshow('src', src)
    cv2.waitKey(0)



cv2.imshow('gray', gray)
cv2.imshow('binary', binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 다각형 근사 근데 시ㅡ발 뭔말인지 못알아 먹겠음..

src = cv2.imread('image/phone.png',1)

gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU)
binary = cv2.bitwise_not(binary)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_TC89_KCOS)

for contour in contours:
    epsilon = cv2.arcLength(contour, True) * 0.02
    approx_poly = cv2.approxPolyDP(contour, epsilon, True)
    
    for approx in approx_poly:
        cv2.circle(src, tuple(approx[0]),3,(255,0,0),2)
        
cv2.imshow('src',src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 코너검출 = 꼭짓점

src = cv2.imread("image/coffee.jpg")
src = cv2.resize(src, dsize = (0,0), fx = 0.5, fy = 0.4, interpolation= cv2.INTER_AREA)
dst = src.copy()

gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 5, blockSize=3, useHarrisDetector=True, k=0.03)

for i in corners:
    cv2.circle(dst, tuple(i[0].astype(np.uint16)), 3, (0, 0, 255), 2, cv2.LINE_AA)
    
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 블록껍질 = 볼록한 부분을 제외하고 윤곽선 검출

src = cv2.imread('image/convex.jpg')
dst = src.copy()

gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)


for i in contours:
    hull = cv2.convexHull(i, clockwise=True)
    cv2. drawContours(dst, [hull], 0, (0,0,255), 2)


cv2.imshow("binary", binary)
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 모멘트, 0차 모멘트부터 3차모멘트 까지 계산 1공간,중심,질량 등 계산가능


src = cv2.imread('image/convex.jpg')
##src = cv2.resize(src, dsize = (0,0), fx = 0.5, fy = 0.4, interpolation= cv2.INTER_AREA)
dst = src.copy()

gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

for i in contours:
    M = cv2.moments(i,False)
    cX = int(M['m10'] / M['m00'])
    cY = int(M['m01'] / M['m00'])
    
    cv2.circle(dst, (cX,cY),3,(255,0,0),-1)
    cv2.drawContours(dst, [i], 0, (0,0,255),2)

cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 모폴로지 변환

src = cv2.imread('image/zebra.jpg',1)
src = cv2.resize(src, dsize=(0,0), fx = 0.3, fy = 0.3, interpolation = cv2.INTER_AREA)

kernel1 = cv2.getStructuringElement(cv2.MORPH_CROSS, (4,4))
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (4,4))
kernel3 = cv2.getStructuringElement(cv2.MORPH_RECT, (4,4))



dilate1 = cv2.dilate(src, kernel1, anchor=(-1,-1), iterations = 5)
erode1 = cv2.erode(src, kernel1, anchor=(-1,-1),iterations = 5)
dilate2 = cv2.dilate(src, kernel2, anchor=(-1,-1), iterations = 5)
erode2 = cv2.erode(src, kernel2, anchor=(-1,-1),iterations = 5)
dilate3 = cv2.dilate(src, kernel3, anchor=(-1,-1), iterations = 5)
erode3 = cv2.erode(src, kernel3, anchor=(-1,-1),iterations = 5)

cv2.imshow('src', src)
cv2.imshow('erode1', erode1)
cv2.imshow('dilate1', dilate1)
cv2.imshow('erode2', erode2)
cv2.imshow('dilate2', dilate2)
cv2.imshow('erode3', erode3)
cv2.imshow('dilate3', dilate3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 모폴로지 연산 

src = cv2.imread('image/office.jpg',1)
src = cv2.resize(src, dsize = (0,0), fx = 0.5, fy = 0.5, interpolation = cv2.INTER_AREA)
src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
dst = cv2.morphologyEx(src, cv2.MORPH_HITMISS, kernel, iterations = 9)

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
